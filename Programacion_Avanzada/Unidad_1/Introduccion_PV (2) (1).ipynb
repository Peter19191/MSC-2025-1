```python
def obtener_radios():
    while True:
        try:
            entrada = input("Ingrese los siguientes valore de radios positivos separados en comas: ")
            radios = [float(r) for r in entrada.split(",") if float(r) > 0]  # Convierte y filtra negativos
            if not radios:
                raise ValueError("Debe ingresar al menos un número real positivo.")
            return radios
        except ValueError:
            print("Error: Asegúrese de ingresar solo números reales positivos.")

radios = obtener_radios()
```

```python
import math

# Calcular el área y perímetro para cada radio ingresado
resultados = [(r, math.pi * r**2, 2 * math.pi * r) for r in radios]
```

```python
# Mostrar resultados
for r, area, perimetro in resultados:
    print(f"\nPara un radio de {r}:")
    print(f"  - El área es: {area:.2f} unidades cuadradas")
    print(f"  - El perímetro es: {perimetro:.2f} unidades")
```
